```python
# import relevant stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from PIL import Image, ImageTk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import mplcursors
```

<h4>The Stickerposters</h4>
Below the part of the dashboard where you're able to select a colour sticker. When selecting a colour, the points where these coloured stickers are places are emphasized.

```python
#https://stackoverflow.com/questions/57970378/tkinter-how-to-display-image-and-not-move-any-content-around
root = tk.Tk()
root.title("De Stickerposters")

# Add a grid
mainframe = tk.Frame(root)
mainframe.grid(column=0,row=0, sticky="NE" )
mainframe.columnconfigure(0, weight = 1)
mainframe.rowconfigure(0, weight = 1)
mainframe.pack(pady = 100, padx = 100)

# Create a Tkinter variable
tkvar = tk.StringVar(root)

# Dictionary with options
choices = {'normaal','groen','rood','blauw','geel'}
tkvar.set('normaal') # set the default option

#the dropdown
popupMenu = tk.OptionMenu(mainframe, tkvar, *choices)
tk.Label(mainframe, text="Kies een kleur").grid(row = 1, column = 2)
popupMenu.grid(row = 2, column =2)

#get legenda in there
legenda = tk.Label(mainframe, image = None)
legenda.grid(row = 4, column = 2)
imgleg = Image.open("stickerposters/legenda.jpg")
imgleg = imgleg.resize((240,250))
photoleg = ImageTk.PhotoImage(imgleg)
legenda.image = photoleg
legenda.configure(image= photoleg)
#get scale in there
scale = tk.Label(mainframe, image = None)
scale.grid(row = 5, column = 2)
imgscale = Image.open("stickerposters/schaal.jpg")
imgscale = imgscale.resize((240,250))
photoscale = ImageTk.PhotoImage(imgscale)
scale.image = photoscale
scale.configure(image= photoscale)


# on change dropdown value
def change_dropdown(*args):
    #get the picture name via row
    kleur = tkvar.get()

    for i in range(4):
        label = i
        label = tk.Label(mainframe, image = None)
        
        if i==0:
            label.grid(row = 4, column = 0)
        elif i==1:
            label.grid(row = 4, column = 1)
        elif i==2:
            label.grid(row = 5, column = 0)
        elif i==3:
            label.grid(row = 5, column = 1)
        img = Image.open("stickerposters/sticker_"+kleur+str(i)+".jpg")
        img = img.resize((240,250))
        photo = ImageTk.PhotoImage(img)
        label.image = photo
        label.configure(image= photo)

        

# link function to change dropdown
tkvar.trace('w', change_dropdown)

root.mainloop()
```

<h4>Getting userprofile</h4>

here someone can select a user to see what they gathered.

```python
#get the relevant datasets
pics = pd.read_csv("fotos.csv", sep = ';')
par = pd.read_csv("participants.csv", sep = ';')
par, pics
```

```python
#https://stackoverflow.com/questions/57970378/tkinter-how-to-display-image-and-not-move-any-content-around
root = tk.Tk()
root.title("De inwoners")

# Add a grid
mainframe = tk.Frame(root)
mainframe.grid(column=0,row=0, sticky="NE" )
mainframe.columnconfigure(0, weight = 1)
mainframe.rowconfigure(0, weight = 1)
mainframe.pack(pady = 100, padx = 100)

# Create a Tkinter variable
tkvar = tk.StringVar(root)

# Dictionary with options
choices = {'Gert','Sasha','Lieve'}
tkvar.set('-') # set the default option

#the dropdown
popupMenu = tk.OptionMenu(mainframe, tkvar, *choices)
tk.Label(mainframe, text="Kies een inwoner").grid(row = 1, column = 0)
popupMenu.grid(row = 2, column = 0)


# on change dropdown value
def change_dropdown(*args):
    #get name of participant, with name id
    inw = tkvar.get()
    inw_id = par.loc[par['Naam'] == inw, 'User']
    
    #get the quote from the participant
    for val in par.loc[par['Naam'] == inw, 'Quote']:
        quote = val

    #show name and quote
    tk.Label(mainframe, text = "Naam: "+inw).grid(row = 4, column = 0)
    tk.Label(mainframe, text = "Wat hij/zij ons heeft verteld: '"+quote+"'").grid(row = 5, column = 0)

    #get pictures that belong with participant
    pic = pics.loc[pics['user'] == int(inw_id), 'foto']
    
    p = 0
    
    #make everything around the image white
    #https://stackoverflow.com/questions/44231209/resize-rectangular-image-to-square-keeping-ratio-and-fill-background-with-black/44231784
    def make_square(im, min_size=250, fill_color=(255, 255, 255, 0)):
        x, y = im.size
        size = max(min_size, x, y)
        new_im = Image.new('RGBA', (size, size), fill_color)
        new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
        return new_im
    
    basewidth = 170
    
    #instert the pictures submitted by the participant
    for i in pic:
        label = i
        label = tk.Label(mainframe, image = None)
        label.grid(row = 6, column = p)
        img = Image.open("foto/"+i)
        wpercent = (basewidth / float(img.size[0])) #https://opensource.com/life/15/2/resize-images-python
        hsize = int((float(img.size[1]) * float(wpercent)))
        img = img.resize((basewidth, hsize))
        #adding white
        new_img = make_square(img)
        #showing photo
        photo = ImageTk.PhotoImage(new_img)
        label.image = photo
        label.configure(image= photo)
        
        #add the explanation to the picture
        for val in pics.loc[pics['foto']== i, 'uitleg']:
            expl = val
        tk.Label(mainframe, text = 'Uitleg: '+expl).grid(row = 7, column = p)
        
        p = p + 1
        #instert placeholder picture to cover-up previous pics
        if p==pic.shape[0]:
            for c in range(4):
                if p>=pic.shape[0] and p<4:
                    label = tk.Label(mainframe, image = None)
                    label.grid(row = 6, column = p)
                    img = Image.open("foto/bewoonder.jpg")
                    wpercent = (basewidth / float(img.size[0])) #https://opensource.com/life/15/2/resize-images-python
                    hsize = int((float(img.size[1]) * float(wpercent)))
                    img = img.resize((basewidth, hsize))
                    #adding white
                    new_img = make_square(img)
                    #showing photo
                    photo = ImageTk.PhotoImage(new_img)
                    label.image = photo
                    label.configure(image= photo)
                    tk.Label(mainframe, text = '                                     ').grid(row = 7, column = p)
                    p = p + 1
                else:
                    break

    #inster the personal stickerposters
    for i in range(4):
        label = i
        label = tk.Label(mainframe, image = None)

        if i==0:
            label.grid(row = 8, column = 0)
        elif i==1:
            label.grid(row = 8, column = 1)
        elif i==2:
            label.grid(row = 8, column = 2)
        elif i==3:
            label.grid(row = 8, column = 3)
        j = 1 + i
        img = Image.open("stickerposters/sticker_par"+str(int(inw_id))+str(j)+".jpg")
        img = img.resize((240,250))
        photo = ImageTk.PhotoImage(img)
        label.image = photo
        label.configure(image= photo)
        
        

# link function to change dropdown
tkvar.trace('w', change_dropdown)

root.mainloop()
```

```python
#https://stackoverflow.com/questions/57970378/tkinter-how-to-display-image-and-not-move-any-content-around
root = tk.Tk()
root.title("De vragen en foto's")

# Add a grid
mainframe = tk.Frame(root)
mainframe.grid(column=0,row=0, sticky="NE" )
mainframe.columnconfigure(0, weight = 1)
mainframe.rowconfigure(0, weight = 1)
mainframe.pack(pady = 100, padx = 100)

# Create a Tkinter variable
tkvar = tk.StringVar(root)

#let's list the quesions
tk.Label(mainframe, text="1. Wat schreeuwt voor jou Eindhoven?").grid(row = 1, column = 0)
tk.Label(mainframe, text="2. Hoe ziet een data-stroom eruit?").grid(row = 2, column = 0)
tk.Label(mainframe, text="3. waarvan zou jij graag data verzameld zien worden?").grid(row = 3, column = 0)
tk.Label(mainframe, text="4. Wat zou een sensor moeten zien?").grid(row = 4, column = 0)
tk.Label(mainframe, text="5. Wat in Eindhoven wil jij graag benadrukken?").grid(row = 5, column = 0)

# Dictionary with options
choices = {'1','2','3', '4', '5'}
tkvar.set('-') # set the default option

#the dropdown
popupMenu = tk.OptionMenu(mainframe, tkvar, *choices)
tk.Label(mainframe, text="Kies een vraag").grid(row = 6, column = 0)
popupMenu.grid(row = 7, column = 0)


# on change dropdown value
def change_dropdown(*args):
    #get the picture name via row
    row = pics.loc[pics['vraag'] == int(tkvar.get()), 'foto']

    p = 0
    
    basewidth = 170

    #make everything around the image white
    #https://stackoverflow.com/questions/44231209/resize-rectangular-image-to-square-keeping-ratio-and-fill-background-with-black/44231784
    def make_square(im, min_size=250, fill_color=(255, 255, 255, 0)):
        x, y = im.size
        size = max(min_size, x, y)
        new_im = Image.new('RGBA', (size, size), fill_color)
        new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
        return new_im

    #how many images are there?
    plt_cnt = row.shape[0]
    #if there are o answers to the question, print
    if plt_cnt == 0:
        for c in range(4):
            if p>=row.shape[0] and p<4:
                label = tk.Label(mainframe, image = None)
                label.grid(row = 8, column = p)
                
                #resizing
                img = Image.open("foto/bewoonder.jpg")
                wpercent = (basewidth / float(img.size[0])) #https://opensource.com/life/15/2/resize-images-python
                hsize = int((float(img.size[1]) * float(wpercent)))
                img = img.resize((basewidth, hsize))
                #adding white
                new_img = make_square(img)
                #showing photo
                photo = ImageTk.PhotoImage(new_img)
                label.image = photo
                label.configure(image= photo)
                tk.Label(mainframe, text = '                                     ').grid(row = 9, column = p)
                p = p + 1
        tk.Label(mainframe, text="Voor deze vraag zijn helaas geen foto's ingezonden.").grid(row = 10, column = 1)
    
    #instert the pictures submitted for this question
    for i in row:
        #coverup error
        tk.Label(mainframe, text = '                                                                                         ').grid(row = 10, column = p)
        
        label = i
        label = tk.Label(mainframe, image = None)
        label.grid(row = 8, column = p)
        img = Image.open("foto/"+i)
        wpercent = (basewidth / float(img.size[0])) #https://opensource.com/life/15/2/resize-images-python
        hsize = int((float(img.size[1]) * float(wpercent)))
        img = img.resize((basewidth, hsize))
        new_img = make_square(img)
        #showing photo
        photo = ImageTk.PhotoImage(new_img)
        label.image = photo
        label.configure(image= photo)
        
        #add the explanation to the picture
        for val in pics.loc[pics['foto']== i, 'uitleg']:
            expl = val
        tk.Label(mainframe, text = 'Uitleg: '+expl).grid(row = 9, column = p)
        
        p = p + 1
        #instert placeholder picture to cover-up previous pics
        if p==row.shape[0]:
            for c in range(4):
                if p>=row.shape[0] and p<4:
                    label = tk.Label(mainframe, image = None)
                    label.grid(row = 8, column = p)
                    img = Image.open("foto/bewoonder.jpg")
                    wpercent = (basewidth / float(img.size[0])) #https://opensource.com/life/15/2/resize-images-python
                    hsize = int((float(img.size[1]) * float(wpercent)))
                    img = img.resize((basewidth, hsize))
                    new_img = make_square(img)
                    #showing photo
                    photo = ImageTk.PhotoImage(new_img)
                    label.image = photo
                    label.configure(image= photo)
                    tk.Label(mainframe, text = '                                     ').grid(row = 9, column = p)
                    p = p + 1
                else:
                    break
        

# link function to change dropdown
tkvar.trace('w', change_dropdown)

root.mainloop()
```

<h4>The final screen: the stickerposter in detail</h4>

```python
#https://stackoverflow.com/questions/57970378/tkinter-how-to-display-image-and-not-move-any-content-around
root = tk.Tk()
root.title("De Stickerposters Detail")

# Add a grid
mainframe = tk.Frame(root)
mainframe.grid(column=0,row=0, sticky="NESW" )
mainframe.columnconfigure(0, weight = 1)
mainframe.rowconfigure(0, weight = 1)
#mainframe.pack(pady = 100, padx = 100)

# Create a Tkinter variable
tkvar = tk.StringVar(root)

# Dictionary with options
choices = {'Participatie','Dataverzameling','Beleidsvorming', 'Transparantie'}
tkvar.set('-') # set the default option

#the dropdown
popupMenu = tk.OptionMenu(mainframe, tkvar, *choices)
tk.Label(mainframe, text="Kies een poster onderwerp").grid(row = 1, column = 0)
popupMenu.grid(row = 2, column = 0)


# on change dropdown value
def change_dropdown(*args):
    #get the poster name
    poster = tkvar.get()
    
    #select the right database for the poster
    if poster == 'Participatie':
        df = pd.read_csv("locatie_puntenposter_Participatie.csv", sep = ';')
    elif poster == 'Dataverzameling':
        df = pd.read_csv("locatie_puntenposter_Data Verzameling.csv", sep = ';')
    elif poster == 'Beleidsvorming':
        df = pd.read_csv("locatie_puntenposter_Beleidsvorming.csv", sep = ';')
    elif poster == 'Transparantie':
        df = pd.read_csv("locatie_puntenposter_Transparantie.csv", sep = ';')

    #create the base for the plot
    figure = plt.Figure(figsize=(7,7))
    ax = figure.add_subplot()
    ax.set(xlim=(-10, 10), ylim=(-10, 10))
    chart_type = FigureCanvasTkAgg(figure, master = root)
    chart_type.draw()
    chart_type.get_tk_widget().grid(row = 3, column = 0)
    
    #create the frame we used on the posters globally
    ax.axhline(0, color = 'black', linewidth = 0.5)
    ax.axvline(0, color = 'black', linewidth = 0.5)

    ax.text(0, -10, "enthousiast", color="black", fontsize=14, ha = 'center', va = 'top')
    ax.text(0, 10, "onverschillig", color="black", fontsize=14, ha = 'center', va = 'top')
    ax.text(-10, 0, "dit herken ik in Eindhoven", color="black", fontsize=14, ha = 'right', 
            va = 'center', rotation = 'vertical')
    ax.text(10, 0, "dit herken ik niet in Eindhoven", color="black", fontsize=14, ha = 'left', 
            va = 'center', rotation = 'vertical')
    
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    
    ax.set_title(poster, fontsize=24)

    #the plot
    ax.scatter(df["x"], df["y"], s = 200, c = df["kleur"])
    
    #show the name of the user with the datapoint
    mplcursors.cursor(ax).connect("add", lambda sel: sel.annotation.set_text(
            par['Naam'][par[df['user'][sel.target.index] == par['User']].index[0]]))
        
    plt.show()
    
        

# link function to change dropdown
tkvar.trace('w', change_dropdown)

root.mainloop()
```
